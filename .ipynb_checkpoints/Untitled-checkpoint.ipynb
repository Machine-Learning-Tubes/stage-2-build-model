## Import Packages

```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt

from sklearn.feature_selection import mutual_info_classif
%matplotlib inline
```

## Initialize Datasets

```python
data_raw = pd.read_csv('datasets/EDA_df.csv')
```

```python
data_raw.head()
```

```python
data_raw.drop(['customer_id', 'tenure', 'churn'], inplace=True, axis=1)
data_raw.head()
```

```python
X = data_raw.iloc[:, 1:20]
Y = data_raw.iloc[:, 0:1]

importance = mutual_info_classif(X, np.ravel(Y))
feat_imp = pd.Series(importance, data_raw.columns[0:len(data_raw.columns) - 1])
feat_imp.sort_values(ascending=True).plot(kind='barh', color = 'teal')
plt.show()
```

Dari data yang dilihat pada plot diatas dapat disimpulkan bahwa yang paling berpengaruh terhadap penggunaan Tellco menurut gender adalah fitur multiple lines yang diberikan oleh Tellco untuk dapat menggunakan Telco secara bersamaan dengan yang lainnya.

```python
X = data_raw.iloc[:, 1:20]
Y = data_raw.iloc[:, 0:1]
importance = mutual_info_classif(X, np.ravel(Y))
feat_imp = pd.Series(importance, data_raw.columns[0:len(data_raw.columns) - 1])
feat_imp.sort_values(ascending=True).plot(kind='barh', color = 'teal')
plt.show()
```

Plot yang ditampilkan menunjukkan bahwa alasan customer lebih memilih untuk berlangganan terhadap servis internet milik Telco karena payment method yang disediakan oleh Telco untuk membayarnya
